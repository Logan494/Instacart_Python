```python
plt.figure(figsize=(8, 6))
age_group_histogram = df_age_group_sample['age_group'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_group_histogram.png'))
plt.show()
```

```python
#linechart for DOW and age
plt.figure(figsize=(10, 5))

sns.lineplot(data = df_age_group_sample, 
             x='orders_day_of_week', 
             y = 'prices', 
             hue='age_group', 
             palette = 'RdYlGn',
             ci=None
            )

plt.title("Weekly Orders by Age Group")
plt.xlabel("Days of the Week")
plt.ylabel("")
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_group_DOW_line.png'))
plt.show()
```

```python
#linechart for HOD and age
plt.figure(figsize=(10, 5))

sns.lineplot(data = df_age_group_sample, 
             x='order_hour_of_day', 
             y = 'prices', 
             hue='age_group', 
             palette = 'RdYlGn',
             ci=None
            )

plt.title("Hourly Orders by Age Group")
plt.xlabel("Hours of The Day")
plt.ylabel("")
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Age_group_HOD.png'))
plt.show()
```

```python
#linechart for DOW and age
plt.figure(figsize=(10, 5))

sns.lineplot(data = df_age_group_sample, 
             x='orders_day_of_week', 
             y = 'prices', 
             hue='age_group', 
             palette = 'RdYlGn',
             ci=None
            )

plt.title("Weekly Orders by Age Group")
plt.xlabel("Days of the Week")
plt.ylabel("")
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'order_age_line.png'))
plt.show()
```

```python
#regions by age
plt.figure(figsize=(10, 6))
sns.countplot(data=df_age_group_sample, x='Region', hue='age_group')

plt.title('Age Group Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Count')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_region.png'))
plt.show()
```

```python
#regions by orders
plt.figure(figsize=(12, 6))
sns.barplot(data=df_age_group_sample, x='Region', y='prices', hue='age_group', ci=None)

plt.title('Average Prices by Age Group and Region')
plt.xlabel('Region')
plt.ylabel('Average Prices')
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_region_orders.png'))
plt.show()
```

```python
# Group by order_id and calculate the sum of prices for each order
order_prices = df_age_group_sample.groupby('order_id')['prices'].sum()

# Merge the aggregated order_prices back to the original DataFrame
df_with_order_prices = df_age_group_sample.merge(order_prices, on='order_id', suffixes=('', '_sum'))

# Calculate the average order cost
average_order_cost = df_with_order_prices.groupby('age_group')['prices_sum'].mean().reset_index()

# Create a line chart for average order cost by age group
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_order_cost, x='age_group', y='prices_sum', marker='o')

# Add labels and title
plt.title('Average Order Cost by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Order Cost')

# Save the figure
plt.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'age_group_order_cost_line.png'))

# Show the plot
plt.show()
```
